```python
import requests
import pandas

pandas.options.display.max_rows = None

url = "https://api.foursquare.com/v3/places/search"

headers = {
    "accept": "application/json",
    "Authorization": "fsq3BH6Xgu8yXWF/gmIZt9XMPIKGcbomsLsd7JaICTkQNKQ="
}

params = {
    "query": 'garage',
    "sort": 'DISTANCE',
    "near": 'Harare'
}

response = requests.get(url, headers=headers, params=params)
```

```python
uk_results = response.json()
```

```python
pandas.json_normalize(uk_results)
```

```python

uk_data_centres = uk_results['results']

d_to_add = []

pandas.options.display.max_colwidth = 50

for res in uk_data_centres:
    for cat in res['categories']:
        cat['id'] = res['fsq_id']
        cat['geocodes'] = res['geocodes']
        cat['address'] = res['location']
        d_to_add.append(cat)

uk_df = pandas.json_normalize(d_to_add)
```

```python
import folium
```

```python
df = pandas.DataFrame(uk_df)

# coordinate for each and every found place
coordinates = []

for index, row in df.iterrows():
    ll_dict = {
        'name' : row['name'],
        'latitude' : row['geocodes.main.latitude'],
        'longitude' : row['geocodes.main.longitude'],
        'address' : row['address.address']
    }
    coordinates.append(ll_dict)
co_normalized = pandas.DataFrame(coordinates)
```

```python
main_longitude = '-17.8162414'
main_latitude = '30.8917018'
m = folium.Map(location=(main_longitude,main_latitude))

for co in coordinates:
    folium.Marker(
    location=[co['latitude'],co['longitude']],
    popup=co['name'],
    icon=folium.Icon(icon="cloud"),
    ).add_to(m)
```
